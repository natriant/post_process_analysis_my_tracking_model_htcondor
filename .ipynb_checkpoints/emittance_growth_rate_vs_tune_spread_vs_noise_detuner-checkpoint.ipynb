### Compute emittance

```python
# #--- <x^2> ---
def mean2(numb):
    return np.mean( (numb - np.mean(numb))**2 )

# #--- <xx'> ---
def mean3(numbx , numbpx):
    return np.mean( (numbx - np.mean(numbx)) * (numbpx - np.mean(numbpx)) )

# #--- sqrt(<x^2> * <px^2> - <xx'>^2) --- compute statistical emittance
def cmp_emit(position, angle):
    return  np.sqrt(mean2(position) * mean2(angle) - mean3(position,angle)**2) # geometrical emittance
```

### Loop over three versions and the values of the noise Delta

```python
versions = [1]#, 2, 3]
Deltas_list = [0.025, 0.05, 0.1]  
```

### Loop over studies. Dictionaries to store the emittance values over turns, 1 dictionay for each version, 1 for each value of delta

```python
ex_dictionary = {}
for version in versions:
    ex_dictionary['v{}'.format(version)] = {}
    for Delta in Deltas_list:
        ex_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)] = {}
        for i, detuner_strength in enumerate(detuner_strengths_list):
            print(Qx_list[i])
            print(version, Delta,detuner_strength)
            ex_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)] = []
            df = pd.read_pickle(my_afs_path+'Qx{:.3f}_detuner{:.2f}_g0.200_delta{:.3f}_betax115.75_particles15000_turns1000_aperture0.03.pkl'.format(Qx_list[i], detuner_strength, Delta))#, version))
            for turn in range(data['study_parameters']['turns']):
                ex_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)].append(cmp_emit(df.at[turn, 'x']/sqrt(data['machine_parameters']['beta_x']), df.at[turn, 'px']*sqrt(data['machine_parameters']['beta_x'] )))
```

### Plot rate vs noise strength vs version
- The color code indicate different versions
- Do a linear fit on the emittance values for each value of Delta, to obtain the emittance growth rate.
- The rate for each vale and run/version is also saved for future use

```python
fig, ax = plt.subplots()
turns_list = np.arange(0, data['study_parameters']['turns'])

colors = ['b', 'r', 'g']
markers = ['o', 's', '*']
color_pointer = 0

rate_dictionary = {} # this dicitonary stores the emittance growth rate for every noise (Delta) value for each one of the different runs (versions)
for version in versions:
    rate_dictionary['v{}'.format(version)] = {}
    for Delta in Deltas_list:
        rate_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)] = {}
        for detuner_strength in detuner_strengths_list:
            rate_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)] = []


for version in versions:
    for index_2, detuner_strength in enumerate(detuner_strengths_list):
        for index, Delta in enumerate(Deltas_list):
            
        # remove the first 200 turns from emittance_x dictionary
            current_study =  ex_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)][0:]
            # Do the linear fit
            [m_h,b_h], cov_h = np.polyfit(turns_list[0:], current_study/current_study[0], 1, cov= True)
            err_h = np.sqrt(np.diag(cov_h))
            rate_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)].append(m_h)
            if detuner_strength == detuner_strengths_list[0]:
                ax.scatter(Qx_list[index_2], m_h, s=40, c = colors[color_pointer], marker = markers[index], label = 'v{}'.format(version))
            else:
                ax.scatter(Qx_list[index_2], m_h, s=40, c = colors[color_pointer], marker = markers[index])
    color_pointer = color_pointer+1
plt.legend()
ax.set_ylim(10e-8, 10e-1)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Linear tune shift '+r'$\xi$')
ax.set_ylabel(r'$\frac{1}{\epsilon_0} \frac{d\epsilon}{dn}$', fontsize = 25)
```

### Plot rate vs noise vs theoretical rate + error bars
- add error bars, standard error $\sigma_x$

\begin{equation}
\sigma_x = \frac{\sigma}{\sqrt{n}}
\end{equation}
, where $\sigma$ is the standard deviation of the population and $n$ the number of observations of the sample

```python
fig, ax = plt.subplots()
turns_list = np.arange(0, data['study_parameters']['turns'])
markers_list = ['v', 'o', 's'] 

A = 2 # A is chosen "by the eye". If you want to calculate it properly, follow the procedure used in "emittance_growth_rate_vs_noise_octupole.ipynb"
for version in versions:
    for index, Delta in enumerate(Deltas_list):     
    
        theory_rate = [] # list to store the theory rate for each values of xi. 0 for every teration over Delta
        for index_2, detuner_strength in enumerate(detuner_strengths_list): 
            if index_2 == 0:
                ax.scatter(my_tune_spreads[index_2],rate_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)], c ='k', marker=markers_list[index] ,label=r'$\Delta={:.3f}$'.format(Delta))
            else:
                ax.scatter(my_tune_spreads[index_2],rate_dictionary['v{}'.format(version)]['Delta{}'.format(Delta)]['detuner_strength{}'.format(detuner_strength)], c ='k', marker=markers_list[index])
            theory_rate.append(A*((my_tune_spreads[index_2]**2)*((Delta**2))/(2*data['Feedback']['values_limits']['max']**2)))
        ax.plot(my_tune_spreads, theory_rate, c = 'k')


plt.axvline(x=0.03, linestyle = '--', c = 'k', label=r'$\xi={}$'.format(0.03))   

ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Linear tune shift ')
ax.set_ylabel(r'$\frac{1}{\epsilon_0} \frac{d\epsilon}{dn}$', fontsize = 40)

# Set the ticks in both sides of the axes
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
# Display minor ticks on the axis
ax.minorticks_on() 

# Set the appearance of the ticks
ax.tick_params(which='both',direction='in', width=2)
ax.tick_params(which='major',direction='in', length=11)
ax.tick_params(which='minor',direction='in', length=4)


# Make axis equal. Note that the size of the figure affects this
ax.set_aspect(aspect='auto')


plt.xlim(4e-3, 0.6)
plt.ylim(5e-7, 7e-2)
savefig = False
plt.legend()
plt.tight_layout()
if savefig: 
    plt.savefig('./figures/rate_vs_noise_vs_Qxvarious_detunerScan_g0.200_deltaScan_betax115.75_particles15000_turns1000_aperture0.03_vertical_line.png')
```
